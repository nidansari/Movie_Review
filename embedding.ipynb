```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
import tensorflow as tf
print(tf.__version__)
```

```python
## Sentences

sent=[ 'The glass of milk',
      'the cup of tea',
      'the glass of juice',
      'I am a good girl',
      'I am a good developer',
      'I love Codingg',
      'Letssgooo'
]
```

```python
sent
```

```python
voc_size= 10000
```

```python
one_hot_repr=[one_hot(words,voc_size) for words in sent]
```

```python
# like Here (the) word basically comed on 6974 index based on vector representation in 6974 index it will get 1 and rest will be 0.


one_hot_repr

```

```python
## Word embedding representation

from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.utils import pad_sequences
```

```python
import numpy as np
```

```python
# Here we are doing padding because we need to give index of same size or input then only model works 

sent_lenght=8
embedded_docs=pad_sequences(one_hot_repr,padding='pre',maxlen=sent_lenght)
print(embedded_docs)
```

```python
# feature representation

dim=10
```

```python
model=Sequential()
model.add(Embedding(voc_size,dim,input_length=sent_lenght))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
model.predict(embedded_docs)
```
